```python
# Load the dataset
df = pd.read_csv('/content/BPD_Arrests - Copy.csv')
(df.head())
```

```python
df.info()
```

```python
# removing the null values
removenull_df = df.dropna(subset=['Latitude', 'Longitude'])
```

```python
#Create a base map centered around Baltimore, Maryland
map_Boltimore = [39.2904, -76.6122]  # Center of Baltimore
m = folium.Map(location=map_Boltimore, zoom_start=11)
```

```python
# Add heatmap for females
df_female = removenull_df[removenull_df['Gender'] == 'Female']
female_data = removenull_df[['Latitude', 'Longitude']].values.tolist()
```

```python
# Add heatmap for males
df_male= removenull_df[removenull_df['Gender'] == 'Male']
male_data = removenull_df[['Latitude', 'Longitude']].values.tolist()
```

```python
# Add a HeatMap layer to the map
HeatMap(male_data).add_to(m)

HeatMap(female_data).add_to(m)
```

```python
# Save the map to an HTML file to view
map_path = "Combined_heatmap_BD4.html"
m.save(map_path)
```

##Key Findings
Male arrests are mainly concentrated in urban and high-risk areas, often linked to crime and poverty. Female arrests are more scattered, particularly in residential neighborhoods related to domestic violence. The overlap between the two suggests that some areas experience crime affecting both genders.

##Potential Insights

The findings can help improve policing strategies, focusing on high-crime areas for men and providing more support services, like shelters, for women. Policies could also aim to reduce gender-based disparities in arrests.

The map highlights possible biases in the criminal justice system and points to the need for reforms. It suggests that interventions should address both gender-specific crime patterns and work toward more equitable treatment for men and women.
