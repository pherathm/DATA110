**1. Import Libraries and Load the Dataset**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data_url = "https://covid.ourworldindata.org/data/owid-covid-data.csv"
df = pd.read_csv(data_url)
```

**2.Data Inspection and Cleaning**

I will inspect the dataset for missing values, identify columns with null entries, and summarize the missing data. Then, I will drop columns with more than 50% missing values. For the remaining missing values, I will replace numerical ones with the median and fill categorical ones with "Unknown" to maintain consistency and clarity.Finally, I will save the cleaned dataset as a new file for further analysis.

```python
# Display the first few rows
print("Dataset Preview:")
print(df.head())
```

```python
# Inspect for missing values
print(df.isnull().sum())
```

```python
# Drop columns with more than 50% missing data
threshold = len(df) * 0.5
df = df.loc[:, df.isnull().sum() <= threshold]

# Fill missing values with median
df.fillna(df.median(numeric_only=True), inplace=True)

# Fill missing values for categorical columns with "Null"
categorical_cols = df.select_dtypes(include=['object']).columns
df[categorical_cols] = df[categorical_cols].fillna("Null")


# Verify the cleaning process and see the data structure
print(df.info())
```

```python
# Save cleaned dataset for further use
df.to_csv("cleaned_covid_dataset.csv", index=False)
```

##**1.Scatter Plot: Total Cases vs. Total Deaths by Country**

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Load the dataset
url = "https://covid.ourworldindata.org/data/owid-covid-data.csv"
data = pd.read_csv(url)

# Filter for relevant data (latest available for each country)
latest_data = data.sort_values(by='date').groupby('location').last().reset_index()

# Filter data for visualization examples
latest_data = latest_data[latest_data['iso_code'].str.len() == 3]  # Valid ISO codes

# Scatter Plot - Total Cases vs. Total Deaths
fig_scatter = px.scatter(
    latest_data,
    x="total_cases",
    y="total_deaths",
    color="continent",
    size="population",
    hover_name="location",
    title="Scatter Plot: Total Cases vs. Total Deaths",
    labels={"total_cases": "Total Cases", "total_deaths": "Total Deaths"}
)
fig_scatter.show()
```

This scatter plot explores the relationship between the total number of COVID-19 cases and deaths reported across various countries. Generally, countries with a higher total number of cases exhibit a corresponding increase in total deaths, indicating a correlation between the two variables. However, certain data points deviate significantly from this trend, reflecting differences in healthcare infrastructure, the quality of interventions such as vaccinations and lockdowns, and variations in reporting standards. Additionally, the use of bubble size to represent population highlights how countries with larger populations contribute more significantly to global case and death statistics.



#**2.Line Chart - Global New Cases Over Time**

```python
global_cases = data.groupby('date')['new_cases'].sum().reset_index()
fig_line = px.line(
    global_cases,
    x="date",
    y="new_cases",
    title="Line Chart: Global New Cases Over Time",
    labels={"new_cases": "New Cases", "date": "Date"}
)
fig_line.show()


```

The line chart illustrates the progression of new COVID-19 cases globally over time, showcasing trends and significant events during the pandemic. Peaks in the chart likely align with major outbreaks or the emergence of variants such as Delta and Omicron. In contrast, periods of decline signal successful containment measures, including lockdowns, widespread vaccination campaigns, and the potential development of herd immunity. Policymakers can use these trends to analyze the effectiveness of interventions and adjust strategies accordingly during different phases of the pandemic.



#**3.Histogram - Distribution of Total Cases**


```python
fig_histogram = px.histogram(
    latest_data,
    x="total_cases",
    nbins=30,
    title="Histogram: Distribution of Total Cases",
    labels={"total_cases": "Total Cases"}
)
fig_histogram.show()
```

This histogram presents the distribution of total COVID-19 cases across countries. The data reveals that most countries have a relatively low total case count, as indicated by the high frequency of cases in the lower ranges. However, there is a long tail in the distribution, with a few countries reporting exceptionally high case numbers. These outliers may be due to factors such as larger populations, significant outbreaks, or differences in testing and reporting. This visualization underscores disparities in the spread and impact of the pandemic globally, shaped by varying public health policies, testing rates, and population density.

#**4. Bar Chart - Top 10 Countries by Total Vaccinations**

```python
top_vaccinated = latest_data.nlargest(10, "total_vaccinations")
fig_bar = px.bar(
    top_vaccinated,
    x="location",
    y="total_vaccinations",
    title="Bar Chart: Top 10 Countries by Total Vaccinations",
    labels={"location": "Country", "total_vaccinations": "Total Vaccinations"}
)
fig_bar.show()
```

The bar chart highlights the top 10 countries that have administered the most COVID-19 vaccinations. These countries are often high-income nations with advanced healthcare systems, better vaccine availability, and infrastructure to support mass immunization efforts. However, the disparities between these top-performing countries and others underscore challenges such as vaccine accessibility, distribution logistics, and public acceptance. These leading countries play a significant role in contributing to global vaccination efforts, which have been critical in reducing the severity and spread of the pandemic.

#**5. Treemap - Vaccination Share by Continent**

```python
fig_treemap = px.treemap(
    latest_data,
    path=["continent", "location"],
    values="total_vaccinations",
    title="Treemap: Vaccination Share by Continent",
    labels={"total_vaccinations": "Total Vaccinations"}
)
fig_treemap.show()
```

The treemap provides a visual representation of the distribution of vaccinations across continents and countries. Larger rectangles represent higher vaccination totals, emphasizing regions and countries leading the global vaccination effort. Within continents, the differences between countries highlight the inequalities in vaccine distribution and accessibility. This visualization draws attention to the dominant role of specific nations in achieving vaccination milestones and underscores the need for equitable vaccine distribution to combat the pandemic effectively.

#**6.Choropleth Map - Vaccinations per 100 People**

```python
fig_choropleth = px.choropleth(
    latest_data,
    locations="iso_code",
    color="total_vaccinations_per_hundred",
    hover_name="location",
    title="Choropleth Map: Vaccinations per 100 People",
    color_continuous_scale=px.colors.sequential.Plasma,
    labels={"total_vaccinations_per_hundred": "Vaccinations per 100 People"}
)
fig_choropleth.show()
```

The choropleth map uses varying shades to depict vaccination rates per 100 people across different countries. Lighter colors indicate higher vaccination rates, often corresponding to nations with robust vaccination campaigns. Conversely, darker shades signify countries with low vaccination rates, possibly due to challenges such as access, hesitancy, or demographic factors. This map effectively highlights global disparities in vaccination efforts, helping identify regions that require greater attention and support to achieve widespread immunity.



#**6.Sankey Diagram: Simulated Flow of Cases**

```python
sankey_data = {
    "source": ["Infected", "Infected", "Recovered", "Recovered", "Deaths"],
    "target": ["Recovered", "Deaths", "Hospitalized", "Quarantined", "Buried"],
    "value": [100, 50, 30, 20, 50]
}
fig_sankey = go.Figure(
    go.Sankey(
        node=dict(
            pad=10,
            thickness=20,
            line=dict(color="black", width=0.5),
            label=["Infected", "Recovered", "Deaths", "Hospitalized", "Quarantined", "Buried"]
        ),
        link=dict(
            source=[0, 0, 1, 1, 2],  # Index of sources
            target=[1, 2, 3, 4, 5],  # Index of targets
            value=sankey_data["value"]
        )
    )
)
fig_sankey.update_layout(title_text="Sankey Diagram: Simulated Flow of Cases", font_size=10)
fig_sankey.show()
```

The Sankey diagram represents a hypothetical flow of COVID-19 cases from initial infection to eventual outcomes, such as recovery or death. The width of each flow indicates the proportion of cases transitioning between states, offering insights into critical stages like hospitalization and quarantine. The majority of cases lead to recovery, while a smaller proportion results in deaths, reflecting the outcomes of interventions and healthcare responses. Although simulated, the diagram mirrors real-world patterns and highlights the importance of effective measures in reducing severe outcomes and managing the disease effectively.
