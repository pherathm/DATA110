### 8. Customizing Scatter Plots

```python
#Generate data for the function f(x)=xsin(x), over the range [0,4pi]
x = np.linspace(0, 4 * np.pi, 100)
y = x * np.sin(x)

#Create the plot
plt.plot(x, y, color ='red', linewidth =2, label = 'line plot')
plt.scatter(x, y, s = 70, color = 'black', label = 'scatter plot', facecolors='orange', edgecolors='black', linewidths=2)

#labels
plt.title('Scatter Plot of f(x) = x * sin(x)')
plt.xlabel('x')
plt.ylabel('f(x)')


```

###9. Combine Plot

```python
# x values from -2pi to 2pi
x = np.linspace(-2 * np.pi, 2 * np.pi, 400)

# f(x) = x * cos(x)
y = x * np.cos(x)

#plot the scatter plot with intervals every pi/4
plt.plot(x,y, label = 'line', color = 'black', linestyle = '-')
x_scatter = np.arange(-2*np.pi, 2*np.pi, np.pi/4)
y_scatter = x_scatter * np.cos(x_scatter)
plt.scatter(x_scatter,y_scatter, label = 'Points', color = 'red')

#labels for the graph
plt.xlabel('x')
plt.ylabel('xcos(x)')
plt.title('graph of xcos(x)')
plt.legend()
```

###10. 10. Custom Plot with Alternative Style

```python
#choosed functions
#1.g(x) = xe^(-x^2)
#2.h(x)=sin(x)/x

# Generate x values from -10 to 10
x = np.linspace(-10, 10, 400)



#Generate x values from -10  to 10
#functions
g = x * np.exp(-x**2) #g(x) = xe^(-x^2)
h = np.sin(x)/ x  #h(x)=sin(x)/x
h[x == 0] = 1  # Set h(0) = 1 manually since it's not defined


plt.plot(x,g, linestyle='-', color='red', linewidth=1, label="g(x) = x * e^(-x^2)")
plt.plot(x, h, linestyle='--', color='green', linewidth=1, label="h(x) = sin(x)/x")

# Add scatter points in h(x)
scatter_x = np.array([-10, -5, 0, 5, 10])
plt.scatter(scatter_x, np.sinc(scatter_x / np.pi), color='black', s=10, label="Scatter Points h(x)")

#Add lables
plt.xlabel('x')
plt.ylabel('g(x) and h(x)')

# Add a legend
plt.legend()

# alternative style
plt.style.use('seaborn-darkgrid')

```

###11. Style Comparison

```python
print(plt.style.available)

#plot the graph
plt.plot(x, x**2 -2*x + 1)

#Style
plt.style.use('seaborn-whitegrid')

#xlabel and ylabel
plt.ylabel('f(x)')
plt.xlabel('x')
plt.title('Style - seaborn-whitegrid')


```

```python
#plot the graph
plt.plot(x, x**2 -2*x + 1)

#Style
plt.style.use('dark_background')

#xlabel and ylabel
plt.ylabel('f(x)')
plt.xlabel('x')
plt.title('Style - dark_background')
```

```python
#plot the graph
plt.plot(x, x**2 -2*x + 1)

#Style
plt.style.use('classic')

#xlabel and ylabel
plt.ylabel('f(x)')
plt.xlabel('x')
plt.title('Style - classic')
```

###Differences
- seaborn-whitegrid: Light, non-intrusive grid, white background, focused on readability.
- dark_background: Sharp contrasting gridlines against the dark background.
- classic: No gridlines with with black axes. Look like  traditional scientific plots.
