###Crime Data Visualization (2022-2023)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("ggplot")
```

###Task 1: Basic Bar Chart for Crime Rates in 2023

```python
#set the data to a data table
crime_data = { 'Offense': ['Aggravated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape', 'Robbery'],
        'Offenses per 100,000 (2023)': [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5],
         'Change in Offenses per 100,000 (2022-2023)': [-3.3, -8.1, -4.9, 12.0, -12.0, -9.8, -8.0] }
df = pd.DataFrame(crime_data)

# Customize colors for each bar
colors = ['blue', 'red', 'green', 'yellow', 'purple', 'lightgreen', 'lightskyblue']

# Bar Chart for Crime Rates in 2023 with customized colors
plt.figure(figsize=(10, 6))
bars = plt.bar(df['Offense'], df['Offenses per 100,000 (2023)'], color=colors)

# Labels and title
plt.xlabel('Offense')
plt.ylabel('Offenses per 100,000 (2023)')
plt.title('Crime Rates per 100,000 People in 2023')

# Gridlines
plt.grid(axis='y', linestyle='-', alpha=0.8)

plt.tight_layout()
plt.show()
```

###Task 2: Change in Crime Rates (2022 vs. 2023) with Unique Colors

```python
# Customize colors for each bar
colors = ['blue', 'red', 'green', 'yellow', 'purple', 'lightgreen', 'lightskyblue']

# Horizontal Bar Chart for Crime Rates in 2023 with customized colors
plt.figure(figsize=(10, 6))
bars = plt.barh(df['Offense'], df['Change in Offenses per 100,000 (2022-2023)'], color=colors)

# Label and title
plt.xlabel('Change of Offenses per 100,000 (2022-2023)')
plt.ylabel('Offence')
plt.title('Percentage Change in Crime Rates per 100,000 People in 2022 Vs.2023')

# Gridlines
plt.grid(axis='x', linestyle='-', alpha=0.8)

# Add a vertical line at x=0
plt.axvline(0, color='black', linestyle = '--')

plt.tight_layout()
plt.show()
```

###Why does this bar graph make more sense than Task 1's bar graph?

##### Think about how absolute numbers (like total crime rates) might be misleading compared to percentage changes.

Absolute numbers are the total counts of crimes for a specific year, such as how many murders happened in 2023. While they give us a clear picture of crime for the year 2023, they don't tell us if things are getting better or worse over the time.
Percentage change shows how much crime rates have increased or decreased compared to the previous year.
By looking at percentage changes, we can see whether crime is really decreasing overall or if some types of crime are getting worse, even if the total numbers seem lower. This comparison is crucial for making decisions about public safety.

###Optional Task: Combined Bar Chart for Crime Rates in 2022 and 2023

```python
#calculate the 2022 crime rates based on the percentage changes

df['Offenses per 100,000 (2022)'] = df['Offenses per 100,000 (2023)'] / (1 + df['Change in Offenses per 100,000 (2022-2023)'] / 100)


# Combined Bar Chart for Crime Rates in 2022 and 2023


# Generate indexes for the x-axis
x_indexes = np.arange(len(df['Offense']))

# Set the width for the bars
bar_width = 0.4

# Plot the bars for 2022
plt.bar(x_indexes - bar_width/2, df['Offenses per 100,000 (2022)'], width=bar_width, label='2022', color='blue')

# Plot the bars for 2023
plt.bar(x_indexes + bar_width/2, df['Offenses per 100,000 (2023)'], width=bar_width, label='2023', color='red')

# Customize the x-axis with offense labels
plt.xticks(ticks=x_indexes, labels= df['Offense'], rotation=45, ha='right')


# labels and title
plt.xlabel('Offense Types')
plt.ylabel('Offenses per 100,000 People')
plt.title('Comparison of Crime Rates for 2022 and 2023')

# Add a legend to distinguish between 2022 and 2023
plt.legend()


plt.tight_layout()
plt.show()
```
