## Dataset
You will be working with the 'Iris' dataset, which is available in the seaborn library. This dataset contains measurements of iris flowers, including sepal length, sepal width, petal length, petal width, and species.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")

iris.head()
```

## Tasks

### 1. Data Visualization

#### 1.1 Pair Plot
Create a pair plot of all features in the Iris dataset, color-coding by species. Include an overall title for the plot.

```python
pair_plot = sns.pairplot(iris, hue = "species", markers = ["o", "s", "p" ])

pair_plot.fig.suptitle("Species sepal length", y=1.1)

plt.show()
```

The diagonal density plots show how each feature is distributed, with Setosa being clearly different in petal width. The scatter plots indicate that petal length and petal width do a great job of separating Setosa from the other species, while sepal length and sepal width have a lot of overlap, especially between Versicolor and Virginica. This suggests that petal measurements are better for telling the species apart than sepal measurements. In terms of correlation, petal length and width are strongly related, especially in Versicolor and Virginica, while sepal length and width are only weakly related. As a result, Setosa can be easily identified using both petal and sepal features, but to tell Versicolor and Virginica apart, it’s important to focus on petal measurements because their sepal features are similar.

#### 1.2 Bar Plot
Create a bar plot showing the average sepal length for each iris species.

```python
plt.figure(figsize = (8,5))
sns.barplot(data=iris, x = 'species', y = "sepal_length", color = 'red')
plt.title('Average Sepal Length for each  Iris species')
plt.xlabel('species')
plt.ylabel('average sepal length (cm)')
plt.show()
```

The bar plot shows the average sepal length for each iris species, highlighting clear differences. Virginica has the longest sepal length, followed by versicolor, with setosa having the shortest. This pattern indicates that sepal length can help distinguish setosa from the others and somewhat separates versicolor from virginica.

#### 1.3 Box Plot

```python
sns.boxplot(data=iris, x= "species", y = "sepal_length", color = 'red')
plt.title('Distribution of Petal widths for each Iris species')
plt.xlabel('species')
plt.ylabel('average petal width (cm)')
plt.show()
```

The box plot displays the sepal length distribution for each iris species. Setosa has a smaller, more consistent range, while versicolor and virginica show more variation, with virginica having the widest spread and highest values. This suggests that sepal length differences can help separate setosa from the others, though there’s some overlap between versicolor and virginica.

#### 1.4 Grouped Bar Plot

```python
sns.barplot(data=iris.melt(id_vars = "species", value_vars= ["sepal_length", "petal_length"]), x = "species", y = "value", hue = "variable", errorbar = None)
plt.title("Average Sepal and petal lengths for three species")
plt.xlabel("Species")
plt.ylabel("Length (cm)")
plt.legend(title = "Type of Measurment", labels = ["Sepal_length", "Petal_length"])
plt.show()
```

The bar plot shows the average sepal and petal lengths for each iris species, allowing for easy comparison. Virginica has the longest averages in both categories, followed by versicolor, with setosa having the shortest. The noticeable gap in petal length between setosa and the other two species suggests that petal length could be especially helpful for distinguishing setosa. Overall, both sepal and petal lengths differ among species, but petal length may be the most effective for separation.

#### 1.5 Violin Plot

```python
sns.violinplot(data=iris, x="species", y="sepal_length", color = 'red')
plt.title("Distribution of Sepal Lengths for Each Iris Species")
plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")
plt.show()
```

The violin plot shows the spread and density of sepal lengths for each iris species. Setosa has a smaller range with high density at lower sepal lengths, while versicolor and virginica have wider distributions, with virginica reaching the highest lengths. This plot highlights that setosa consistently has shorter sepal lengths, making it easier to differentiate from the other species. However, there is some overlap between versicolor and virginica, although virginica tends to have larger sepal lengths overall.
